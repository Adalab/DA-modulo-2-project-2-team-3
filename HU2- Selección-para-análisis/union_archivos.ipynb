```python
#Vamos a unir 3/4 archivos dado que en uno de ellos no habia columnas de nuestro interes.

import pandas as pd
```

```python
kaggel_final= pd.read_csv('kaggel_modificado2.csv')
kaggel_final.head()
```

```python
txt_final= pd.read_csv('txt_csv2.csv')
txt_final.head()
```

```python
xml_final= pd.read_csv('xml_csv_2.csv')
xml_final.head()
```

```python
nuevas_columnas = {col: col.lower().replace(" ", "_") for col in kaggel_final.columns}
nuevas_columnas
```

```python
kaggel_final.rename(columns = nuevas_columnas, inplace = True)
kaggel_final.head(1)
```

```python
#PRIMER MERGE:
xml_txt_final=xml_final.merge(txt_final, how='outer', on='Unnamed: 0')

```

```python
xml_txt_final.columns
```

```python
#SEGUNDO MERGE Y FINAL: 
data_final=xml_txt_final.merge(kaggel_final, how='outer', left_on='Unnamed: 0',right_on='unnamed:_0')
```

```python
data_final.to_csv('data_seleccionada.csv')
```
